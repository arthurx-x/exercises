1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
prev = pd.read_csv('previsao_de_renda.csv')
````

````python
def categorical_analysis(variable, title):
    plt.figure(figsize=(7, 6))
    sns.barplot(x=variable, y='renda', data=prev, errorbar='sd', legend=False)
    plt.title(title)
    plt.show()
categorical_analysis('posse_de_veiculo', 'Análise de Renda por Posse de Veículo')
categorical_analysis('posse_de_imovel', 'Análise de Renda por Posse de Imóvel')
plt.figure(figsize=(10, 8))
categorical_vars = ['posse_de_veiculo', 'posse_de_imovel', 'renda']
correlation_matrix = prev[categorical_vars].corr(method='spearman')  # Usando correlação de Spearman para dados categóricos
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlação para Posse de Veículo e Imóvel vs. Renda')
plt.show()
````

Nenhuma das variáveis parece ser uma boa preditora/discriminalizadora de renda de forma estatisticamente significante.
