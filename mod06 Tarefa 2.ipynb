```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nascidos = pd.read_csv('SINASC_RO_2019.csv')
nascidos.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
ax = sns.boxplot(data=nascidos, x='IDADEMAE', color='#179a29')
ax.set_title('Distribuição Idade Mãe', fontsize=16)
ax.set_xlabel('Idade Mãe', fontsize=14)
ax.set_ylabel('Distribuição', fontsize=14)
ax.set_xticks(range(0, 110, 5))
ax.set_xticklabels([str(x) for x in range(0, 110, 5)])
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(12, 8))
ax = sns.boxplot(data=nascidos, x='PESO', color='#179a29')
ax.set_title('Distribuição Peso Bebê ao Nascer', fontsize=16)
ax.set_xlabel('Peso Bebê', fontsize=14)
ax.set_ylabel('Distribuição', fontsize=14)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(12, 8))
ax = sns.histplot(data=nascidos, x='IDADEMAE', bins=30, kde=True, color='#179a29')
ax.set_title('Distribuição Idade Mãe', fontsize=16)
ax.set_xlabel('Idade Mãe', fontsize=14)
ax.set_ylabel('Frequência', fontsize=14)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(12, 8))
ax = sns.histplot(data=nascidos, x='PESO', bins=30, kde=True, color='#179a29')
ax.set_title('Distribuição Peso ao Nascer', fontsize=16)
ax.set_xlabel('Peso Bebê', fontsize=14)
ax.set_ylabel('Frequência', fontsize=14)
plt.show()
```
