````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
prev = pd.read_csv('previsao_de_renda.csv')
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
plt.figure(figsize=(20, 5))
sns.histplot(data=prev, x='data_ref', hue='posse_de_imovel', multiple='stack', shrink=0.8)
plt.title('Distribuição Posse de Imóvel')
plt.show()
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
variaveis_qualitativas = ['posse_de_imovel', 'posse_de_veiculo']
def plot_mean_over_time(data, qualitative_vars, date_col):
    for var in qualitative_vars:
        grouped_data = data.groupby([date_col, var]).size().reset_index(name='counts')
        plt.figure(figsize=(20, 5))
        sns.lineplot(x=date_col, y='counts', hue=var, data=grouped_data, marker='o')
        plt.title(f'Média de {var} ao Longo do Tempo')
        plt.xlabel('Data')
        plt.ylabel('Média')
        plt.legend(title=var)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
plot_mean_over_time(prev, variaveis_qualitativas, 'data_ref')
````

- não parece haver alterações anormais na distribuição da posse de imóvel ao longo do tempo. A constancia é uma caracteristica de variaveis uteis para modelos preditivos, acredito.
