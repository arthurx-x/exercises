#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nascidos = pd.read_csv('SINASC_RO_2019.csv')
nascidos.head()
```

```python
print(nascidos.columns)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(15, 20))

plt.subplot(5, 1, 1)
sns.countplot(data=nascidos, x='APGAR1')
plt.title('Distribuição de APGAR1')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

plt.subplot(5, 1, 2)
sns.countplot(data=nascidos, x='APGAR5')
plt.title('Distribuição de APGAR5')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

plt.subplot(5, 1, 3)
sns.countplot(data=nascidos, x='SEXO')
plt.title('Distribuição de SEXO')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

plt.subplot(5, 1, 4)
sns.countplot(data=nascidos, x='GESTACAO')
plt.title('Distribuição de GESTACAO')
plt.xticks([])
plt.yscale("log")
plt.xlabel('')

plt.subplot(5, 1, 5)
sns.countplot(data=nascidos, x='ESTCIVMAE')
plt.title('Distribuição de ESTCIVMAE')
plt.yscale("log")
plt.xlabel('ESTCIVMAE')

plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=nascidos, x='CONSULTAS', y='APGAR5', color='blue', errorbar=None)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title('APGAR5 Médio por Número de Consultas de Pré-natal')
plt.xlabel('Número de Consultas')
plt.ylabel('Média de APGAR5')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.figure(figsize=(15, 30))

plt.subplot(4, 1, 1)
sns.countplot(data=nascidos, x='SEXO')
plt.title('Distribuição de SEXO')
plt.yscale("log")

plt.subplot(4, 1, 2)
sns.countplot(data=nascidos, x='GESTACAO')
plt.title('Distribuição de GESTACAO')
plt.yscale("log")

plt.subplot(4, 1, 3)
sns.countplot(data=nascidos, x='ESTCIVMAE')
plt.title('Distribuição de ESTCIVMAE')
plt.yscale("log")

plt.subplot(4, 1, 4)
sns.countplot(data=nascidos, x='CONSULTAS')
plt.title('Distribuição de CONSULTAS')
plt.yscale("log")

plt.tight_layout()
plt.show()
```
